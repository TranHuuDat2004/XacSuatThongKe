1. mean(): Trung bình cộng của dữ liệu.

```python
import statistics

data = [10, 20, 30, 40, 50]

# Tính giá trị trung bình
mean_value = statistics.mean(data)

print("Giá trị trung bình:", mean_value)
```

2.	statistics.fmean(): Trung bình cộng với số thập phân, tính nhanh hơn, có thể trọng số tùy chọn

```python
import statistics

# Dữ liệu số
data = [8.0, 7.8, 10, 8.5]

# Trọng số tương ứng (được tính dựa trên phần trăm)
weights = [0.1, 0.2, 0.2, 0.5]

# Tính điểm trung bình nhanh chóng với hàm statistics.fmean()
weighted_average = statistics.fmean(data, weights)

# In kết quả
print("Điểm trung bình:", weighted_average)
```

3.	statistics.geometric_mean(): Trung bình hình học của dữ liệu.

```python
import statistics

# Dữ liệu số
data = [2, 4, 8, 16, 32]

# Tính trung bình hình học với hàm statistics.geometric_mean()
geometric_mean = statistics.geometric_mean(data)

# In kết quả
print("Geometric Mean:", round(geometric_mean))
```

4. statistics.harmonic_mean(): Trung bình điều hoà của dữ liệu.

```python
import statistics

# Dữ liệu số
data = [2, 4, 8, 16, 32]

# Tính trung bình điều hòa với hàm statistics.harmonic_mean()
harmonic_mean = statistics.harmonic_mean(data)

# In kết quả
print("Harmonic Mean:", round(harmonic_mean,2))

```

11. statistics.quantiles(): Chia dữ liệu thành các khoảng có xác suất bằng nhau.

19. Ví dụ code minh họa của lớp statistics.NormalDist

```python
import numpy as np
import statistics

# Doanh số bán hàng hàng ngày của sản phẩm A và B trong một tháng
sales_A = np.array([150, 120, 130, 140,
                    160, 170, 180, 190,
                    200, 210, 220, 230,
                    240, 250, 260, 270,
                    280, 290, 300, 310])

sales_B = np.array([100, 110, 120, 130,
                    140, 150, 160, 170,
                    180, 190, 200, 210,
                    220, 230, 240, 250,
                    260, 270, 280, 290])

# Tính trung bình, phương sai và độ lệch chuẩn của sản phẩm A
mean_A = statistics.fmean(sales_A)
variance_A = statistics.pvariance(sales_A)
stdev_A = statistics.pstdev(sales_A)

# Tính trung bình, phương sai và độ lệch chuẩn của sản phẩm B
mean_B = statistics.fmean(sales_B)
variance_B = statistics.pvariance(sales_B)
stdev_B = statistics.pstdev(sales_B)

print("Sản phẩm A:")
print("Trung bình:", mean_A)
print("Phương sai:", variance_A)
print("Độ lệch chuẩn:", stdev_A)

print("-----------------------------------")
print("Sản phẩm B:")
print("Trung bình:", mean_B)
print("Phương sai:", variance_B)
print("Độ lệch chuẩn:", stdev_B)

```
